1、pd.read_csv()一般是读取csv文件的，pd.read_table()读取txt文件。函数pd.read_csv（）和pd.read_table（）的内容相同，只是默认分隔符不同。

在read_csv（）中，定界符为 “ , ” ，在read_table（）中，定界符为 \ t 。
通过sep=’,’，让pd.read_table()实现pd.read_csv()的功能：同理，通过sep=’\t’，让pd.read_csv()实现pd.read_table()的功能。

2、Python的csv模块准确的讲应该叫做dsv模块，因为它实际上是支持范式的分隔符分隔值文件（DSV，delimiter-separated values）的。
delimiter参数值默认为半角逗号，即默认将被处理文件视为CSV。
当delimiter=’\t’时，被处理文件就是TSV。
